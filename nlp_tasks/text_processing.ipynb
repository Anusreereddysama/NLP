```python
from nltk.tokenize import sent_tokenize,word_tokenize
sentences= sent_tokenize(paragraph)
sentences
```

```python
from nltk.stem import PorterStemmer
Stemming=PorterStemmer()
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

for i in range(len(sentences)):
    words=word_tokenize(sentences[i]) 
    print(f"Original Words in {i}th Sentence : {words}")
    stemming_words=[Stemming.stem(word) for word in words]
    print(f"Stemmed Words in {i}th Sentence : {stemming_words}")
    lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in words]
    print(f"Lemmatized Words in {i}th Sentence : {lemmatized_words}")
words
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=" ".join(words)
```
