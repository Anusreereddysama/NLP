```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```

```python
#Stemming-It is the process of a reducing words to the root stem
words=["eating","eats","eaten","writing","written","programming","programs","history","finalized"]
```

```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
for  word in words:
   print(word+ "---->"+ stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem("sitting")
```

```python
from nltk.stem import RegexpStemmer
regex=RegexpStemmer('ing$|e$|s$|able$|,min=4')
```

```python
regex.stem("eating")
```

```python
regex.stem('ingeating')
```

```python
from nltk.stem import SnowballStemmer
SnowballStemmer =SnowballStemmer('english')
for  word in words:
   print(word+ "---->"+ SnowballStemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
stemming.stem('goed')
```

```python
#Lemmatization : like stemming. The output we will get after lemmatization is called 'lemma',which is the root word rather than root stem.
#Wordnet Lemmatizer

from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize("going",pos='v')
```

```python
for  word in words:
   print(word+ "---->"+ lemmatizer.lemmatize(word,pos='v'))
```

```python
lemmatizer.lemmatize("fairly",pos='v'),lemmatizer.lemmatize("sportingly",pos='v')
```
